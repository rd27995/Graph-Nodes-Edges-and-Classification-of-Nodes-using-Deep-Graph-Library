```python
import matplotlib.pyplot as plt
```

```python
import torch as th
```

```python
import dgl
import numpy as np
```

```python
import networkx as nx
```

```python
g_nx = nx.petersen_graph()
```

```python
g_dgl = dgl.DGLGraph(g_nx)
```

## Visualizing Inbuilt Graphs from Networkx and Converting it to DGL Graph

```python
plt.subplot(121)
nx.draw(g_nx,with_labels=True)

plt.subplot(122)
nx.draw(g_dgl.to_networkx(),with_labels=True)

plt.show()
```

## Using Source and Destination Arrays to create a Graph

```python
u = th.tensor([0,0,0,0,0])
v = th.tensor([1,2,3,4,5])
star1 = dgl.DGLGraph((u,v))

print('We Have {} nodes'.format(star1.number_of_nodes()))
print('We Have {} edges'.format(star1.number_of_edges()))

nx.draw(star1.to_networkx(),with_labels=True)
```

## Using list of edges to create a Graph

```python
edge_list = [(0,1),(1,2),(2,3),(3,1),(2,0),(4,2),(5,3),(5,2),(0,4)]
star2 = dgl.DGLGraph(edge_list)

nx.draw(star2.to_networkx(),with_labels=True)
```

## Progressively adding edges and Nodes

```python
g = dgl.DGLGraph()
g.add_nodes(10)

for i in range(1,4):
    g.add_edge(i,0)


src = list(range(5,8))
dst = [0]*3

g.add_edges(src,dst)

src = th.tensor([8, 9]); dst = th.tensor([0, 0])
g.add_edges(src, dst)

g.clear(); g.add_nodes(10)
src = th.tensor(list(range(1, 10)));
g.add_edges(src, 0)

nx.draw(g.to_networkx(),with_labels=True)
plt.show()
```
